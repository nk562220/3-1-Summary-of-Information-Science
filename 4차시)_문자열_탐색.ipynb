# 1. brute force 법

```python
def brute_force_search(text, pattern):
    n = len(text)
    m = len(pattern)
    total = []

    if m == 0 or n<m: #계산할 필요 없음
        return total

    for i in range(n - m + 1): #n-m 까지 순회, 텍스트 길이가 패턴 보다 짧을 순 없음
        j = 0 #초기화
        while j < m and text[i + j] == pattern[j]:#i 번째 칸부터 텍스트와 패턴의 값 1대1 대조
            j += 1

        if j == m: # 완전, 끝까지 일치시
          total.append(i)


    return total
```

# 라빈 카프 알고리즘

```python
def rabin_karp_search(text, pattern, q=139):
    """
    라빈-카프 알고리즘을 사용하여 텍스트 내에서 패턴을 검색하는 함수입니다.
    매칭되는 모든 시작 인덱스를 리스트로 반환합니다.
    """
    n = len(text)
    m = len(pattern)

    if m == 0 or n < m:
        return []

    d = 2  # 진수

    result = []
    p_hash = 0  # 패턴의 해시값
    t_hash = 0  # 텍스트 부분 문자열의 해시값
    h = 1

    # h = (d^(m-1)) % q 계산
    for i in range(m - 1):
        h = (h * d)

    h = h%q

    # 패턴과 텍스트의 첫 번째 부분 문자열에 대한 초기 해시값 계산
    for i in range(m):
        p_hash = (d * p_hash + ord(pattern[i])) % q
        t_hash = (d * t_hash + ord(text[i])) % q

    print(f"패턴의 해시값: {p_hash}")

    # 텍스트를 순회하며 해시값 비교 및 롤링 해시 갱신
    for i in range(n - m + 1):
        print(f"인덱스 {i}에서의 본문 부분 문자열 해시값: {t_hash}")

        if p_hash == t_hash:
            # 해시값이 일치하면 실제 문자를 하나씩 비교 (해시 충돌 대비)
            match = True
            for j in range(m):
                if text[i + j] != pattern[j]:
                    match = False
                    break
            if match:
                result.append(i)

        # 다음 부분 문자열의 해시값 계산 (롤링 해시)
        if i < n - m:
            # 첫 문자 제거 및 새로운 문자 추가
            t_hash = (((t_hash - ord(text[i]) * h) << 1) + ord(text[i + m])) % q

            # 음수 해시값이 나올 경우 양수로 변환
            if t_hash < 0:
                t_hash += q

    return result
```

# KMP 알고리즘

```python
#KMP 알고리즘

def compute_lps_array(pattern):
#패턴
    m = len(pattern)
    lps = [0] * m
    length = 0  # 이전 최대 접두사 길이
    i = 1 #lsp의 첫 값은 항상 0

    while i < m: #끝까지 순회
        if pattern[i] == pattern[length]: #패턴의 현재 문자[i] == 접두사의 다음 문자(length)
            length += 1 #다음 인덱스 검사
            lps[i] = length
            i += 1
        else:
            if length != 0: #문자 불일치, 하지만 이전에 이미 일치했던 접두사 존재
                length = lps[length - 1] #
            else:
                lps[i] = 0
                i += 1

    print(pattern)
    print(lps)
    return lps

def kmp_search(text, pattern):
    """
    KMP 알고리즘을 사용하여 텍스트 내에서 패턴을 검색하는 함수입니다.
    매칭되는 모든 시작 인덱스를 리스트로 반환합니다.
    """
    n = len(text)
    m = len(pattern)

    if m == 0 or n<m:
        return []

    lps = compute_lps_array(pattern)
    result = []
    i = 0  # 텍스트의 인덱스
    j = 0  # 패턴의 인덱스

    while i < n:
        if pattern[j] == text[i]:
            i += 1
            j += 1

        if j == m:
            # 패턴을 찾은 경우
            result.append(i - j)
            j = lps[j - 1]
        elif i < n and pattern[j] != text[i]:
            # 불일치가 발생한 경우
            if j != 0:
                j = lps[j - 1]
            else:
                i += 1

    return result
```

# 보이어 무어 알고리즘

```python
#보이어 무어 건너뛰기 테이블x 버전

def generate_bad_character_table(pattern):
    """
    패턴 내 각 문자의 가장 오른쪽 등장 위치를 기록하는
    나쁜 문자 테이블을 생성하는 함수입니다.
    """
    m = len(pattern)
    bad_char_table = {}

    for i in range(m):
        bad_char_table[pattern[i]] = i


    print(bad_char_table)
    return bad_char_table

def boyer_moore_search(text, pattern):

    n = len(text)
    m = len(pattern)

    if m == 0 or n <m:
        return []

    bad_char_table = generate_bad_character_table(pattern)
    result = []

    s = 0

    while s <= n - m:
        j = m - 1

        while j >= 0 and pattern[j] == text[s + j]:
            j -= 1

        if j < 0:
            result.append(s)

            if s + m < n:
                bad_char_index = bad_char_table.get(text[s + m], -1)
                s += m - bad_char_index
            else:
                s += 1
        else:
            bad_char_index = bad_char_table.get(text[s + j], -1)
            s += max(1, j - bad_char_index)

    return result
```

```python
#보이어 무어 건너뛰기 테이블 버전

def generate_skip_table(pattern):
    m = len(pattern)
    skip_table = {}

    # 패턴의 마지막 문자를 제외한 나머지 문자들에 대해 건너뛰기 거리 계산
    # 수식: 패턴의 길이(m) - 1 - 현재 인덱스(i)
    for i in range(m - 1):
        skip_table[pattern[i]] = m - 1 - i


    print(skip_table)
    return skip_table

def boyer_moore_search(text, pattern):
    n = len(text)
    m = len(pattern)
    total = []

    if m == 0 or n < m:
        return total

    skip_table = generate_skip_table(pattern)
    s = 0

    while s <= n - m:
        j = m - 1

        # 뒤에서부터 패턴과 텍스트 비교
        while j >= 0 and pattern[j] == text[s + j]:
            j -= 1

        if j < 0:
            # 패턴이 완전히 일치한 경우
            total.append(s)
            # 일치 후 다음 탐색을 위한 이동: 윈도우의 마지막 문자를 기준으로 이동
            s += skip_table.get(text[s + m - 1], m)
        else:
            # 불일치 발생 시: 불일치 위치 j가 아닌, 윈도우의 마지막 텍스트 문자를 기준으로 이동
            s += skip_table.get(text[s + m - 1], m)

    return total
```

# 2번 문제

```python
#2번

def compute_lps_array(pattern):

    m = len(pattern)
    lps = [0] * m
    length = 0
    i = 1

    while i < m:
        if pattern[i] == pattern[length]:
            length += 1
            lps[i] = length
            i += 1
        else:
            if length != 0:
                length = lps[length - 1]
            else:
                lps[i] = 0
                i += 1

    return lps


def min_repeat(s):

    lps = compute_lps_array(s)
    n = len(s)

    # 최소 반복 단위 후보 길이
    repeat_len = n - lps[-1]

    # 전체 문자열이 해당 단위로 정확히 반복되는지 확인
    if n % repeat_len == 0:
        return s[:repeat_len]
    else:
        return s


# 실행 예시
string = input("문자열 입력: ")
repeat = min_repeat(string)
print("최소 반복 단위:", repeat)
```

```python
#3번

def longest_common_substrings(A, B, q):
    left = 0
    right = min(len(A), len(B))
    max_len = 0

    # 1️⃣ 최장 길이 찾기
    while left <= right:
        mid = (left + right) // 2
        found = False

        for i in range(len(A) - mid + 1):
            sub = A[i:i+mid]
            if rabin_karp_search(B, sub, q):
                found = True
                break

        if found:
            max_len = mid
            left = mid + 1
        else:
            right = mid - 1

    # 2️⃣ 모든 부분 문자열 찾기
    result_set = set()

    for i in range(len(A) - max_len + 1):
        sub = A[i:i+max_len]
        if rabin_karp_search(B, sub, q):
            result_set.add(sub)

    return max_len, result_set


# 🔽 실행 부분
print("⚠️ 영문(알파벳)만 입력하세요. 한글 입력 금지")

A = input("문자열 A 입력: ")
B = input("문자열 B 입력: ")
q = int(input("소수 입력: "))

length, substrings = longest_common_substrings(A, B, q)

print(f"최장 길이: {length}")

if length == 0:
    print("가능한 문자열이 없습니다.")
else:
    print("가능한 부분 문자열:")
    for s in sorted(substrings):
        print(s)
```
